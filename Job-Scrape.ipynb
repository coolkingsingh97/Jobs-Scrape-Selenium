```python
from selenium.webdriver import Chrome
import time
#connect to chrome

browser = Chrome()

#open web page

browser.get('https://www.indeed.com/l-Seattle-jobs.html')
time.sleep(10)

# perform navigation clicks to desired page

# browser.find_element_by_xpath('//button[@class="mat-button guided-search-button ng-star-inserted"]').click()
# time.sleep(2)
# browser.find_element_by_xpath('//div[@class="guided-search-option-name" and text()=" Medical Specialties "]').click()
# time.sleep(2)
# browser.find_element_by_xpath('//div[@class="guided-search-option-name" and text()=" Medical Care "]').click()
# time.sleep(2)
# browser.find_element_by_xpath('//div[@class="guided-search-option-name" and text()=" Hospitals "]').click()
# time.sleep(5)
# browser.find_element_by_xpath('//div[@class="dropdown-container applied-filters"]').click()
# time.sleep(2)
# browser.find_element_by_xpath('//span[@class="ng-star-inserted" and text()=" 100 miles " ]').click()


```

```python
Summary= []
JobAddress= []
JobNames= []
CompanyNames=[]




for i in range(1,101):
    
     # Extracting the names of jobs from the page
        
            job_names = browser.find_elements_by_xpath('//a[@class="jobtitle turnstileLink " ]')
            for names in job_names:
                JobNames.append(names.text)

# print(len(JobNames))
        # Extracting the Address of hospitals from the page

            job_address = browser.find_elements_by_xpath('//*[@class="location accessible-contrast-color-location"]')
            for address in job_address:
                JobAddress.append(address.text)
            
#         # Extracting the Ratings of hospitals from the page

            summary = browser.find_elements_by_xpath('//div[@class="summary"]')
            for summ in summary:
                Summary.append(summ.text)
           
            
# Extracting the numbers of hospitals from the page
            company_names = browser.find_elements_by_xpath('//span[@class="company"]')
            for cnames in company_names:
                CompanyNames.append(cnames.text)
        
#         #Break condition to stop at the last page    
#         if i == Pages:
#             break
            
#         # Shifting to the next page
        
            browser.find_element_by_xpath('//span[@class="np" and contains(text(),"Next")]').click()
            time.sleep(8)
            
print(len(Summary))
print(len(JobNames))
print(len(JobAddress))
print(len(CompanyNames))
```

```python
import pandas as pd

# Converting lists to a list of tuples

JobzippedList =  list(zip(JobNames, JobAddress, CompanyNames, Summary))


# Making a DataFrame

Jobs_DF = pd.DataFrame(JobzippedList, columns = ['Position Name' , 'Work_Address', 'Company','Requirements'])


# Splitting the Address column into address,city,state_pin

Jobs_DF[['City','State_Pin']] = Jobs_DF.Work_Address.str.split(",",expand=True) 

print(Jobs_DF)


# Export to excel


Jobs_DF.to_excel("Jobs.xlsx")


```
